## PGF Processing

```python
## Processing MAX Temperature data

# Creating a main dataframe to store data from other csv files

os.chdir(r"D:\TUM\Master_Thesis\Benin\Data\PGF v3\time_series")

df = pd.read_csv(r"D:\TUM\Master_Thesis\Benin\Data\PGF v3\Processed_max\1998-01-01.csv")
row, col = df.shape
print(df.shape)
time_series = pd.DataFrame()
i = 0
j = 0
ls =[]
for r in range(row):
    
    for c in range(col-1):
        
        x = df.iloc[r,0]
        y = df.columns.values[c+1]
        col_head = str(x)+","+str(y) #defining the column name using the lat,long
        time_series[col_head] = df.iloc[r,c+1]
        time_series[col_head] = 'loc_'+str(i)
        i=i+1

time_series.to_csv("main.csv")

##---------------------------------------------------------------------------------------------

#Processing data from other csv files to the main csv file

os.chdir(r"D:\TUM\Master_Thesis\Benin\Data\PGF v3\Processed_max")

i=0

for year in range(1998,2013):
    
    filepath = r'D:\TUM\Master_Thesis\Benin\Data\PGF v3\tmax_daily_'+str(year)+'-'+str(year)+'.nc'
    ds = nc.Dataset(filepath)
    
    time = ds.variables['time'][:]
    
    startdate = "01-01-"+str(year)
    day_counter = 0
    
    print("Processing year:", year)
    
    for day in range(len(time)):
        ls=[]
        
        enddate = pd.to_datetime(startdate) + pd.DateOffset(days = day_counter)
        enddate = enddate.strftime('%Y-%m-%d') #removing the time stamp from date
        
        filename = str(enddate)+".csv"
        
        df = pd.read_csv(filename)

        row, col = df.shape
        
        for r in range(row):
            
            for c in range(col-1):
                
                val = df.iloc[r,c+1]
                ls.append(val)
                
        time_series.loc[i] = ls
        i = i+1 
        day_counter = day_counter+1
        
time_series.to_csv("tmax_1998-2012.csv")
```

## ERA5 Processing

```python
## Processing MAX and MIN Temperature datasets for ERA5 csv files

# Creating a main dataframe to store data from other csv files

os.chdir(r"D:\TUM\Master_Thesis\Benin\Data\ERA5\time_series")

df = pd.read_csv(r"D:\TUM\Master_Thesis\Benin\Data\ERA5\csv_files\1998-01-01_max.csv")
row, col = df.shape

time_series = pd.DataFrame()

i = 0
j = 0
ls =[]
for r in range(row):
    
    for c in range(col-1):
        
        x = df.iloc[r,0]
        y = df.columns.values[c+1]
        col_head = str(x)+","+str(y) #defining the column name using the lat,long
        time_series[col_head] = df.iloc[r,c+1]
        time_series[col_head] = 'loc_'+str(i)
        i=i+1

time_series.to_csv("main.csv")

##---------------------------------------------------------------------------------------------##

#Function defined to process data from other csv files to the main csv file

def ERA5timeseries(ending):

    os.chdir(r"D:\TUM\Master_Thesis\Benin\Data\ERA5\csv_files")

    i=0

    day_start = date(1998,1,1)
    day_end = date(2012,12,31)

    time = day_end - day_start
    startdate = '01-01-1998'
    day_counter = 0

    for day in range(time.days+1):
        ls = []
   
        enddate = pd.to_datetime(startdate) + pd.DateOffset(days = day_counter)
        enddate = enddate.strftime('%Y-%m-%d') #removing the time stamp from date
        print("Processing Date", enddate)
        
        file = str(enddate) + "_" + ending + ".csv"
        
        df = pd.read_csv(file)

        row1, col1 = df.shape
        
        for r in range(row1):
            
            for c in range(col1-1):
                
                val = df.iloc[r, c + 1] 
                ls.append(val)
        
        time_series.loc[i] = ls
        
        i = i + 1 
        day_counter = day_counter + 1
        
    time_series.to_csv(ending+"_1998-2012.csv")
```

```python
#Calling the function

ERA5timeseries("max")
ERA5timeseries("min")
```

## EWEMBI Processing

```python
## Processing of EWEMBI csv files

# Creating a main dataframe to store data from other csv files

os.chdir(r"D:\TUM\Master_Thesis\Benin\Data\EWEMBI_v1.1-second_download\time_series")

df = pd.read_csv(r"D:\TUM\Master_Thesis\Benin\Data\EWEMBI_v1.1-second_download\csv_files\1998-01-01_max.csv")
row, col = df.shape

time_series = pd.DataFrame()

i = 0
j = 0
ls =[]
for r in range(row):
    
    for c in range(col-1):
        
        x = df.iloc[r,0]
        y = df.columns.values[c+1]
        col_head = str(x)+","+str(y) #defining the column name using the lat,long
        time_series[col_head] = df.iloc[r,c+1]
        time_series[col_head] = 'loc_'+str(i)
        i=i+1

time_series.to_csv("main.csv")

##---------------------------------------------------------------------------------------------##

#Defining the function to convert csv files to time series data

def EWEMBItimeseries(ending):
    os.chdir(r"D:\TUM\Master_Thesis\Benin\Data\EWEMBI_v1.1-second_download\csv_files")

    i=0

    day_start = date(1998,1,1)
    day_end = date(2012,12,31)

    time = day_end - day_start
    startdate = '01-01-1998'
    day_counter = 0

    for day in range(time.days+1):
        ls = []
   
        enddate = pd.to_datetime(startdate) + pd.DateOffset(days = day_counter)
        enddate = enddate.strftime('%Y-%m-%d') #removing the time stamp from date
        print("Processing Date", enddate)
        
        file = str(enddate) + "_" + ending + ".csv"
        
        df = pd.read_csv(file)

        row1, col1 = df.shape
        
        for r in range(row1):
            
            for c in range(col1-1):
                
                val = df.iloc[r, c + 1] 
                ls.append(val)
        
        time_series.loc[i] = ls
        
        i = i + 1 
        day_counter = day_counter + 1
        
    time_series.to_csv(ending + "_1998-2012.csv")
```

```python
#calling the function

EWEMBItimeseries("max")
EWEMBItimeseries("min")
```
